### Question 5.Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(0)

# Generate synthetic data
n_samples = 100
x1 = np.random.rand(n_samples)
x2 = x1 + np.random.normal(0, 0.1, n_samples)  # x2 is correlated with x1
x3 = np.random.rand(n_samples)
x4 = 0.5 * x1 + 0.5 * x3 + np.random.normal(0, 0.1, n_samples)  # x4 is correlated with both x1 and x3

# Create a DataFrame
data = pd.DataFrame({
    'Feature 1': x1,
    'Feature 2': x2,
    'Feature 3': x3,
    'Feature 4': x4
})

# Calculate the correlation matrix
correlation_matrix = data.corr()

# Visualize the correlation matrix using a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm', square=True)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

### Question 4.Using the given x and y data, generate a 3D surface plot to visualize the function

        x, y = np.meshgrid(x, y)
        z = np.sin(np.sqrt(x**2 + y**2))

        'X': x.flatten(),
        'Y': y.flatten(),

        }

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generate the x and y data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

# Create a DataFrame
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}

df = pd.DataFrame(data)

# Create a 3D surface plot
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
ax.plot_surface(df['X'].values.reshape(x.shape), 
                df['Y'].values.reshape(y.shape), 
                df['Z'].values.reshape(z.shape), 
                cmap='viridis', edgecolor='none')

# Set labels and title
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_zlabel('Z Axis')
ax.set_title('3D Surface Plot of Z = sin(sqrt(X^2 + Y^2))')

# Show the plot
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Set the random seed for reproducibility
np.random.seed(25)

# Create the dataset
data = {
    'Country': ['USA', 'Canada', 'UK', 'Germany', 'France'],
    'Population': np.random.randint(100, 1000, 5),
    'GDP': np.random.randint(500, 2000, 5)
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Create the bubble chart
plt.figure(figsize=(10, 6))
plt.scatter(df['GDP'], df['Population'], 
            s=df['Population']*10,  # Bubble size (scaled for visibility)
            alpha=0.6, 
            c=df['Population'], 
            cmap='viridis', 
            edgecolors='w', 
            linewidth=0.5)

# Annotate points with country names
for i in range(len(df)):
    plt.annotate(df['Country'][i], 
                 (df['GDP'][i], df['Population'][i]), 
                 fontsize=9, 
                 ha='center')

# Set labels and title
plt.title('Bubble Chart of Country Population vs GDP')
plt.xlabel('GDP (in billions)')
plt.ylabel('Population (in millions)')
plt.colorbar(label='Population Size')  # Color bar for population sizes
plt.grid(True)
plt.show()
```

### Question 1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Generate x values and compute y values (sine of x)
x = np.linspace(0, 10, 100)  # 100 points from 0 to 10
y = np.sin(x)

# Output to the notebook
output_notebook()

# Create a Bokeh plot
p = figure(title='Sine Wave', x_axis_label='x', y_axis_label='sin(x)', width=800, height=400)

# Add a line renderer for the sine wave
p.line(x, y, legend_label='sin(x)', line_color='blue', line_width=2)

# Show the plot
show(p)
```

### Question 2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns

```python
import numpy as np
import pandas as pd
from bokeh.plotting import figure, show
from bokeh.io import output_file

# Set the output to an HTML file
output_file("scatter_plot.html")

# Generate random data
np.random.seed(42)  # For reproducibility
num_points = 100
x = np.random.rand(num_points) * 100  # Random x values
y = np.random.rand(num_points) * 100  # Random y values
sizes = np.random.rand(num_points) * 50 + 10  # Random sizes (10 to 60)
colors = np.random.rand(num_points, 3)  # Random colors (RGB)

# Create a DataFrame for better organization (optional)
data = {
    'x': x,
    'y': y,
    'sizes': sizes,
    'colors': [f"rgb({int(r*255)}, {int(g*255)}, {int(b*255)})" for r, g, b in colors]
}
df = pd.DataFrame(data)

# Create a Bokeh scatter plot
p = figure(title='Random Scatter Plot', x_axis_label='X', y_axis_label='Y', width=800, height=400)

# Add circle markers with varying sizes and colors
p.scatter('x', 'y', size='sizes', color='colors', source=df, fill_alpha=0.6)

# Show the plot
show(p)
```

### Question 4.Create a Bokeh histogram to visualize the distribution of the given data.
     data_hist = np.random.randn(1000)
    hist, edges = np.histogram(data_hist, bins=30)

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_file

# Set the output to an HTML file
output_file("histogram.html")

# Generate random data
data_hist = np.random.randn(1000)

# Compute the histogram
hist, edges = np.histogram(data_hist, bins=30)

# Create a Bokeh figure
p = figure(title='Histogram of Random Data', 
           x_axis_label='Value', 
           y_axis_label='Frequency', 
           width=800, 
           height=400)

# Add a quad renderer for the histogram
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], 
       fill_color='blue', line_color='black', alpha=0.7)

# Show the plot
show(p)
```

### Question 5. Create a Bokeh heatmap using the provided dataset.
    data_heatmap = np.random.rand(10, 10)
    x = np.linspace(0, 1, 10)
    y = np.linspace(0, 1, 10)
    xx, yy = np.meshgrid(x, y)

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_file
from bokeh.transform import linear_cmap
from bokeh.models import ColorBar
from bokeh.layouts import column
from bokeh.palettes import Viridis256

# Set the output to an HTML file
output_file("heatmap.html")

# Generate random data
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Prepare data for the heatmap
data = {
    'x': xx.flatten(),
    'y': yy.flatten(),
    'value': data_heatmap.flatten()
}

# Create a Bokeh figure
p = figure(title='Heatmap', 
           x_axis_label='X', 
           y_axis_label='Y', 
           width=600, 
           height=600)

# Create a color map
mapper = linear_cmap(field_name='value', palette=Viridis256, low=data['value'].min(), high=data['value'].max())

# Add rectangles to the heatmap
p.rect(x='x', y='y', width=0.1, height=0.1, source=data, line_color=None, fill_color=mapper)

# Create a color bar
color_bar = ColorBar(color_mapper=mapper['transform'], width=8, location=(0,0))
p.add_layout(color_bar, 'right')

# Show the plot
show(p)
```
